```python
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import selenium.webdriver as webdriver
import time

with webdriver.Chrome(service=Service(ChromeDriverManager().install())) as driver:
    driver.get("http://live.xacte.com/lamarathon")
    WebDriverWait(driver, 10).until(EC.visibility_of_element_located((By.XPATH, "//option[@value='2426']"))).click()
    WebDriverWait(driver, 10).until(EC.visibility_of_element_located((By.XPATH, "//option[@value='100']"))).click()
    
    time.sleep(5)

    while True:
        for row in WebDriverWait(driver, 10).until(EC.visibility_of_all_elements_located((By.XPATH, "//table[@id='xact_results_search_results']//tr[.//td]"))):
            data = [cell.text for cell in row.find_elements(By.CSS_SELECTOR,"td")]
            print(data)

        try:
            WebDriverWait(driver, 10).until(EC.visibility_of_element_located((By.XPATH, "//a[@id='xact_results_search_results_next'][.='Next']"))).click()
            WebDriverWait(driver,10).until(EC.staleness_of(row))
        except Exception as err:
            break
```
